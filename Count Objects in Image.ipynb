```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox
from numpy.lib.polynomial import poly
```

```python
img = cv2.imread('data/Delhi_Traffic_EPS.jpg')
box, label, count = cv.detect_common_objects(img)
result = draw_bbox(img, box, label, count)
plt.imshow(result)
plt.show()
```
